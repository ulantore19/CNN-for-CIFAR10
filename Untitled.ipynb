# The implementaition of CNN for CIFAR-10

```python
from torch.nn import Module
from torch.nn import Conv2d
from torch.nn import Linear
from torch.nn import MaxPool2d
from torch.nn import ReLU
from torch.nn import LogSoftmax
from torch.nn import Flatten
import torchvision.transforms as transforms
import torch
```

```python
# Setting device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# Hyperparameters
input_size = 784
num_classes = 10
batch_size = 128
print(f"Training on {device} ...")
```

```python
from torchvision.transforms import ToTensor 
from torchvision import datasets
from torch.utils.data import DataLoader, Dataset

transform = transforms.Compose([
    ToTensor(), 
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Laoding dataset
trainset = datasets.CIFAR10(root='./data', train=True, 
                            download=True, transform=transform)

testset = datasets.CIFAR10(root='./data', train=False, 
                           download=True, transform=transform)



classes = ('plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

```python
# trainset = CIFAR10Dataset(trainset.data, trainset.targets)
# testset = CIFAR10Dataset(testset.data, testset.targets)


# Separating trainset to train and val sets
from torch.utils.data import random_split


train_split = 0.75
numTrainSamples = int(len(trainset) * train_split)
numValSamples = len(trainset) - numTrainSamples

(trainset, valset) = random_split(trainset, (numTrainSamples, numValSamples),
                                  generator=torch.Generator().manual_seed(42))

trainloader = DataLoader(trainset, batch_size=batch_size, shuffle=True, num_workers=2)
valloader = DataLoader(valset, batch_size=batch_size, shuffle=True, num_workers=2)
testloader = DataLoader(testset, batch_size=batch_size,
                                          shuffle=True, num_workers=2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Visualization of the input data

def image_show(image):
    image = image / 2 + 0.5
    # Transpose convert [channels, width, height] -> [height, width, channels]
    plt.imshow(np.transpose(image, (1, 2, 0)))

rand_index = np.random.randint(8)
images, labels = next(iter(trainloader))
images = images.numpy()
fig = plt.figure(figsize=(25, 4))

N = 20  # No. of images to display 
for idx in range(N):
    ax = fig.add_subplot(2, 10, idx+1)
    image_show(images[idx])
    ax.set_title(classes[labels[idx]])
```

```python
class LeNet(Module):
    def __init__(self, input_channels, num_classes):
        super(LeNet, self).__init__()
        self.conv1 = Conv2d(in_channels=input_channels, out_channels=20, kernel_size=(5, 5))
        self.pool = MaxPool2d(kernel_size=(2, 2), stride=(2, 2))
        self.drop = torch.nn.Dropout(0.3)
        self.conv2 = Conv2d(in_channels=20, out_channels=50, kernel_size=(5, 5))
        self.fc1 = Linear(in_features=1250, out_features=500)
        self.fc2 = Linear(in_features=500, out_features=num_classes)
        
    def forward(self, x):
        x = ReLU()(self.conv1(x))
        #x = self.drop(x)
        x = self.pool(x)
        x = ReLU()(self.conv2(x))
        #x = self.drop(x)
        x = self.pool(x)
        x = Flatten()(x)
        x = ReLU()(self.fc1(x))
        x = self.drop(x)
        output = self.fc2(x)
        return output
```

```python
learning_rate = 1e-3
num_epochs = 30
```

```python
model = LeNet(input_channels=3, num_classes=10)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
loss = torch.nn.CrossEntropyLoss()
```

```python
import time
# initialize a dictionary to store training history
H = {
    "train_loss": [],
    "train_acc": [],
    "val_loss": [],
    "val_acc": []
}

# measure how long training is going to take
# print("[INFO] training the network...")
# startTime = time.time()
```

```python
# Define check function to measure validation and test accuracy

def check_acc_loss(model, data_loader, loss_fn, check_classes=False):
    loss_val = 0
    acc = 0
    num_perform_batch = 0
    
    model.eval()
    for idx, (feature, label) in enumerate(data_loader):
        with torch.no_grad():
            feature, label = feature.to(device=device), label.to(device=device)
            scores = model(feature)
            loss_err = loss_fn(scores, label)
            _, pred = scores.max(1)
            
            optimizer.step()

            num_correct = (pred == label).sum()
            num_samples = pred.shape[0]
            #num_samples = batch_size
            acc_batch = num_correct / num_samples
            acc += acc_batch
            loss_val += loss_err
            num_perform_batch += 1
        
        return loss_val / num_perform_batch, acc / num_perform_batch

def performance_check(model, classes, data_loader, loss_fn):
    correct_pred = {classname: 0 for classname in classes}
    total_pred = {classname: 0 for classname in classes}
    with torch.no_grad():
        for features, labels in data_loader:
            features, labels = features.to(device=device), labels.to(device=device)
            _, predictions = torch.max(model(features), 1)
            
            for label, pred in zip(labels, predictions):
                
                if label == pred:
                    correct_pred[classes[label]] += 1
                
                total_pred[classes[label]] += 1
    
    for classname, correct_count in correct_pred.items():
        accuracy = 100 * float(correct_count) / total_pred[classname]
        print(f'Accuracy for class: {classname:5s} is {accuracy:.1f} %')
```

```python
# Model training

for epoch in range(num_epochs):
    
    train_loss = 0
    train_acc = 0
    val_loss = 0
    val_acc = 0
    
    if torch.cuda.is_available():
        model.cuda()
    
    num_perform_batch = 0
    model.train()

    for idx, (train_feature, train_label) in enumerate(trainloader):
        
        train_feature, train_label = train_feature.to(device=device), train_label.to(device=device)
        train_feature = train_feature.cuda()
        scores = model(train_feature)
        loss_err = loss(scores, train_label)
        
        _, pred = scores.max(1)
        
        optimizer.zero_grad()
        loss_err.backward()
        optimizer.step()
        
        num_correct = (pred == train_label).sum()
        num_samples = pred.shape[0]

        acc = num_correct / num_samples
        train_acc += acc
        train_loss += loss_err
        num_perform_batch += 1
    
    avg_train_loss = train_loss / num_perform_batch
    avg_train_acc = train_acc / num_perform_batch
    
    (x_los, x_acc) = check_acc_loss(model, valloader, loss_fn=loss)
    
    H['train_loss'].append(avg_train_loss.cpu().detach().numpy())
    H['train_acc'].append(avg_train_acc)
    H['val_loss'].append(x_los.cpu().detach().numpy())
    H['val_acc'].append(x_acc)
    
    
    print(f"The Epoch({epoch + 1}):\nThe Trainset. loss: {H['train_loss'][-1]}. The accuracy: {H['train_acc'][-1]}")
    print(f"The Validation set. Loss: {H['val_loss'][-1]}. The accuracy: {H['val_acc'][-1]}")
```

```python
test_loss, test_acc = check_acc_loss(model, testloader, loss_fn=loss)
print(f"The test loss is {test_loss.item():.2f}")
print(f"The test accuracy is {test_acc.item():.2f}")
```

Here, we can observe the low validation accuracy (~0.78) and low test accuracy (0.66). Thus, the model is overfitting. In order to solve this problem I will first attempt to make model deeper, meaning that add more layers.

```python
class LeNet(Module):
    def __init__(self, input_channels, num_classes):
        super(LeNet, self).__init__()
        self.conv1 = Conv2d(in_channels=input_channels, out_channels=32, kernel_size=(3, 3))
        self.pool = MaxPool2d(kernel_size=(2, 2), stride=(2, 2))
        self.drop = torch.nn.Dropout(0.5)
        self.conv2 = Conv2d(in_channels=32, out_channels=64, kernel_size=(3, 3))
        self.conv3 = Conv2d(in_channels=64, out_channels=128, kernel_size=(3, 3))
        self.conv4 = Conv2d(in_channels=128, out_channels=256, kernel_size=(3, 3))
        self.fc1 = Linear(in_features=512, out_features=500)
        self.fc2 = Linear(in_features=500, out_features=64)
        self.fc3 = Linear(in_features=64, out_features=num_classes)
        
    def forward(self, x):
        x = ReLU()(self.conv1(x))
        x = self.pool(x)
        x = ReLU()(self.conv2(x))
        x = self.pool(x)
        x = ReLU()(self.conv3(x))
        x = self.pool(x)
        #x = ReLU()(self.conv4(x))
        
        x = Flatten()(x)
        x = ReLU()(self.fc1(x))
        x = self.drop(x)
        x = ReLU()(self.fc2(x))
        x = self.drop(x)
        output = self.fc3(x)
        return output
```

```python
model = LeNet(input_channels=3, num_classes=10)
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
loss = torch.nn.CrossEntropyLoss()
```

```python
H = {
    "train_loss": [],
    "train_acc": [],
    "val_loss": [],
    "val_acc": []
}
```

```python
# Model training

for epoch in range(num_epochs):
    
    train_loss = 0
    train_acc = 0
    val_loss = 0
    val_acc = 0
    
    if torch.cuda.is_available():
        model.cuda()
    
    num_perform_batch = 0
    model.train()

    for idx, (train_feature, train_label) in enumerate(trainloader):
        
        train_feature, train_label = train_feature.to(device=device), train_label.to(device=device)
        train_feature = train_feature.cuda()
        scores = model(train_feature)
        loss_err = loss(scores, train_label)
        
        _, pred = scores.max(1)
        
        optimizer.zero_grad()
        loss_err.backward()
        optimizer.step()
        
        num_correct = (pred == train_label).sum()
        num_samples = pred.shape[0]

        acc = num_correct / num_samples
        train_acc += acc
        train_loss += loss_err
        num_perform_batch += 1
    
    avg_train_loss = train_loss / num_perform_batch
    avg_train_acc = train_acc / num_perform_batch
    
    (x_los, x_acc) = check_acc_loss(model, valloader, loss_fn=loss)
    
    H['train_loss'].append(avg_train_loss.cpu().detach().numpy())
    H['train_acc'].append(avg_train_acc)
    H['val_loss'].append(x_los.cpu().detach().numpy())
    H['val_acc'].append(x_acc)
    
    
    print(f"The Epoch({epoch + 1}):\nThe Trainset. loss: {H['train_loss'][-1]}. The accuracy: {H['train_acc'][-1]}")
    print(f"The Validation set. Loss: {H['val_loss'][-1]}. The accuracy: {H['val_acc'][-1]}")
```

```python
test_loss, test_acc = check_acc_loss(model, testloader, loss_fn=loss)
print(f"The test loss is {test_loss.item():.2f}")
print(f"The test accuracy is {test_acc.item():.2f}")
print("-------------------------------------------")
performance_check(model, classes, testloader, loss_fn=loss)
```
